- 1.Read the file using spark dataframe reader api
- 2.Add metadata columns ->Source File -> Ingestion timestamp
- 3.Write to Bronze delta table

```python
source_file = f"{landing_folder_path}/constructors.json"
table_name = f"{catalog_name}.{bronze_schema_name}.constructors"
```

#Step 1 - Read the file using spark dataframe reader api

```python
#Define schema
constructors_schema = """constructorId STRING,
                         name STRING, 
                         nationality STRING, 
                         url STRING"""
```

```python
#read data from json file
constructors_df = (spark.read
                   .format("json")
                   .schema(constructors_schema)
                   .option('mode', 'FAILFAST')
                   .load(source_file)
                   )
```

```python
display(constructors_df)
```

#Step 2 - Add Metadata

```python
constructors_final_df = add_ingestion_metadata(constructors_df)
```

#Step 3 - Write to bronze delta table

```python
(
    constructors_final_df
    .write
    .mode('overwrite')
    .format('delta')
    .saveAsTable(table_name)
)
```
